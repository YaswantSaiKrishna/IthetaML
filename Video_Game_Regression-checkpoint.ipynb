```python
from sklearn.preprocessing import StandardScaler
```

```python
y=sales.Global_Sales
X=sales.drop('Global_Sales',axis=1)
```

```python
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor

regDTR = DecisionTreeRegressor(random_state = 4)
regDTR.fit(X_train_scaled, y_train)
print("{}".format(regDTR.score(X_train_scaled,y_train)))
print("{}".format(regDTR.score(X_test_scaled,y_test)))
```
